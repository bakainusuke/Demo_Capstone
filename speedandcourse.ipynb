```python



def haversine_distance(lat1, lon1, lat2, lon2):
    """
    Calculate the great circle distance between two points 
    on the earth (specified in decimal degrees)
    """
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])

    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2
    c = 2 * np.arcsin(np.sqrt(a))
    r = 6371  # Radius of earth in kilometers. Use 3956 for miles
    return c * r * 1000  # return distance in meters


def calculate_initial_compass_bearing(lat1, lon1, lat2, lon2):
    """
    Calculate the bearing between two points 
    on the earth (specified in decimal degrees)
    """
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])

    dLon = lon2 - lon1
    y = np.sin(dLon) * np.cos(lat2)
    x = np.cos(lat1) * np.sin(lat2) - np.sin(lat1) * \
        np.cos(lat2) * np.cos(dLon)
    bearing = np.arctan2(y, x)

    # convert from radians to degrees
    bearing = np.degrees(bearing)

    # normalize the bearing to be within 0-360 degrees
    bearing = (bearing + 360) % 360
    return bearing


# Same Haversine distance and initial bearing functions as before


def calculate_speed_and_course(df):
    """Calculates speed and course for each vessel based on MMSI.

    Args:
        df: A pandas DataFrame with columns 'timestamp', 'latitude', 'longitude', and 'mmsi'.
        It is assumed that the 'timestamp' column is of datetime type.

    Returns:
        The input DataFrame with added 'speed' and 'course' columns.
    """

    # Convert timestamp to datetime if it's not already.
    # Ensure to replace the format with your actual format if it's different from '%Y-%m-%d %H:%M:%S UTC'
    if not pd.api.types.is_datetime64_dtype(df['timestamp']):
        df['timestamp'] = pd.to_datetime(
            df['timestamp'], format='%Y-%m-%d %H:%M:%S UTC')

    # Apply calculations for each vessel (grouped by 'mmsi')
    def calculate_for_vessel(vessel_df):
        # Create a mask to identify rows where speed and course are both NaN or null
        mask = vessel_df[['speed', 'course']].isnull().all(axis=1)

        # Calculate distance only for masked rows
        vessel_df.loc[mask, 'distance'] = haversine_distance(
            vessel_df.loc[mask, 'latitude'], vessel_df.loc[mask, 'longitude'],
            vessel_df.loc[mask, 'latitude'].shift(
                -1), vessel_df.loc[mask, 'longitude'].shift(-1)
        )

        # Calculate time difference only for masked rows
        vessel_df.loc[mask, 'time_diff_seconds'] = (
            vessel_df.loc[mask,
                          'timestamp'].shift(-1) - vessel_df.loc[mask, 'timestamp']
        ).dt.total_seconds()
        vessel_df.loc[mask, 'time_diff_seconds'] = vessel_df.loc[mask,
                                                                 'time_diff_seconds'].fillna(0)

        # Calculate speed (in knot) only for masked rows
        vessel_df.loc[mask, 'speed'] = (
            vessel_df.loc[mask, 'distance'] /
            vessel_df.loc[mask, 'time_diff_seconds']
        ) * 1.94384

        # Replace infinite values with NaN (from division by zero)
        vessel_df['speed'] = vessel_df['speed'].replace(
            [np.inf, -np.inf], np.nan)

        # Calculate initial bearing (course) only for masked rows
        vessel_df.loc[mask, 'course'] = calculate_initial_compass_bearing(
            vessel_df.loc[mask, 'latitude'], vessel_df.loc[mask, 'longitude'],
            vessel_df.loc[mask, 'latitude'].shift(
                -1), vessel_df.loc[mask, 'longitude'].shift(-1)
        )

        return vessel_df

    return df.groupby('mmsi').apply(calculate_for_vessel).reset_index(drop=True).fillna(0)
```
